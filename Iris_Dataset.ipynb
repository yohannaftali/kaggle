# Iris Data
Iris adalah genus tumbuhan berbunga dari famili Iridaceae yang terdiri dari sekitar 260 spesies. Tanaman iris biasanya tumbuh dari umbi atau rimpang dan memiliki bunga yang indah dan menarik perhatian. Bunga iris biasanya terdiri dari enam kelopak bunga (tepals) yang berukuran sama dan teratur, yang terdiri dari tiga sepal dan tiga petal, serta tiga benang sari dan satu kepala putik.

"Edgar Anderson's Iris data" adalah kumpulan data yang terkenal di bidang statistik dan pembelajaran mesin, khususnya di bidang pengenalan pola. Kumpulan data ini pertama kali diperkenalkan oleh Edgar Anderson dalam makalahnya yang berjudul "The Species Problem in Iris" yang diterbitkan pada tahun 1936.

>Edgar Anderson (1897-1969) adalah seorang ahli botani dan ahli genetika Amerika yang memberikan kontribusi besar pada bidang sistematis tumbuhan dan biologi evolusi. Ia dikenal sebagai salah satu penemu konsep ecotype, yang menggambarkan populasi tumbuhan yang berbeda yang telah beradaptasi dengan lingkungan tertentu.
Anderson juga merupakan tokoh penting dalam studi hubungan evolusi antara spesies tumbuhan, dan ia adalah pendukung konsep hibridisasi, yang menyatakan bahwa spesies tumbuhan baru dapat muncul melalui persilangan spesies yang ada. Ia memberikan kontribusi besar pada studi genus Quercus (pohon ek) dan kelompok tumbuhan lainnya, dan karyanya membantu membentuk pemahaman kita tentang sejarah evolusi organisme-organisme tersebut.
Secara keseluruhan, Anderson adalah tokoh pionir dalam bidang ilmu tumbuhan, dan karyanya membantu menetapkan banyak konsep dan metode kunci yang masih digunakan dalam bidang ini hingga saat ini.

Kumpulan data ini terdiri dari pengukuran panjang sepal, lebar sepal, panjang kelopak, dan lebar kelopak dari tiga spesies bunga iris yang berbeda: Iris setosa, Iris versicolor, dan Iris virginica. Setiap spesies memiliki 50 sampel, sehingga total terdapat 150 observasi dalam kumpulan data ini. Kumpulan data ini sering digunakan sebagai contoh dalam analisis data eksploratori, visualisasi, dan algoritma pengelompokan seperti K-means.

Iris data adalah sebuah dataset standar dalam dunia Machine Learning yang berisi informasi mengenai tiga jenis bunga iris (Iris setosa, Iris versicolor, dan Iris virginica). Setiap jenis bunga iris dikelompokkan berdasarkan sepal length, sepal width, petal length, dan petal width.

Dataset iris.data terdiri dari 150 baris data, di mana masing-masing baris mewakili satu bunga iris. Setiap baris terdiri dari 5 kolom, yaitu:

- Kolom 1: sepal length dalam satuan cm
- Kolom 2: sepal width dalam satuan cm
- Kolom 3: petal length dalam satuan cm
- Kolom 4: petal width dalam satuan cm

> Sepal adalah bagian dari bunga yang berfungsi sebagai pelindung bagi kelopak bunga (calyx) dan organ reproduksi bunga. Sepal biasanya terdiri dari daun-daun kecil yang berada di bagian luar kelopak bunga dan terletak di pangkal kelopak.

> Pada bunga iris, sepal terletak di bagian luar bunga dan biasanya berwarna hijau atau kecoklatan. Sepal iris seringkali memiliki bentuk dan ukuran yang sama atau hampir sama dengan kelopak bunga (petal), sehingga sulit dibedakan dengan mata telanjang. Namun, dalam analisis data iris, sepal biasanya diukur secara terpisah dengan petal dan digunakan sebagai salah satu fitur dalam klasifikasi dan clustering.

>Petal (atau disebut juga daun mahkota) adalah bagian dari bunga yang berfungsi sebagai bagian yang menarik perhatian serangkaian bunga. Petal biasanya lebih besar dan lebih berwarna-warni dibandingkan dengan sepal dan kelopak bunga.

> Pada bunga iris, petal terletak di bagian dalam bunga dan biasanya memiliki warna yang lebih cerah dan beragam dibandingkan dengan sepal. Petal iris biasanya terdiri dari tiga daun mahkota yang besar dan lebar yang diatur secara radial, dan bisa memiliki warna ungu, putih, biru, atau kuning.



Dataset iris.data sering digunakan sebagai contoh dalam machine learning dan data mining karena sifatnya yang sederhana, mudah dimengerti, dan ukurannya yang kecil. Dataset ini sering digunakan untuk tugas-tugas seperti klasifikasi, regresi, dan clustering.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns;
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
```

Menyiapkan data

```python
# cara 1 menggunakan external data
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
df = pd.read_csv(url, header=None)
```

```python
# cara 2 menggunakan sns dataset
sns.set(style='ticks', color_codes=True)

iris = sns.load_dataset('iris')
iris
```

```python
# cara 3 menggunakan dataset dari library sklearn
iris = load_iris()
iris
```

```python
# ubah menjadi pandas dataset
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = [iris.target_names[i] for i in iris.target]
df
```

Eksplorasi data

```python
sns.pairplot(df, hue='species', height=2, aspect=1)
```

Pisahkan data menjadi fitur dan target

```python
X = df.iloc[:, [0, 1, 2, 3]].values
y = df.iloc[:, -1].values
```

```python
# menampilkan data y
pd.unique(y)
```

Buat model k-means

```python
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=0)
```

```python
y_kmeans = kmeans.fit_predict(X)
```

```python
df['cluster'] = y_kmeans
```

Visualisasikan hasil clustering dengan menggunakan Matplotlib:


```python
sns.scatterplot(data=df, x='sepal length (cm)', y='sepal width (cm)', hue='cluster', palette='bright')
sns.scatterplot(x = kmeans.cluster_centers_[:, 0], y = kmeans.cluster_centers_[:, 1], color='black', marker='X', s=200, label = 'Centroids')
plt.title('Plot Clusters of Iris Dataset (Sepal Features)')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

```python
sns.scatterplot(data=df, x='petal length (cm)', y='petal width (cm)', hue='cluster', palette='bright')
sns.scatterplot(x = kmeans.cluster_centers_[:, 2], y = kmeans.cluster_centers_[:, 3], color='black', marker='X', s=200, label = 'Centroids')
plt.title('Plot Clusters of Iris Dataset (Sepal Features)')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```
