#Unsupervised Learning

Unsupervised learning adalah salah satu jenis pembelajaran mesin di mana algoritma diinstruksikan untuk menemukan pola atau struktur tersembunyi di dalam data tanpa memerlukan label atau jawaban yang benar sebelumnya. Dalam unsupervised learning, model diberikan kumpulan data mentah tanpa label dan mempelajari karakteristiknya sendiri.

Terdapat dua jenis unsupervised learning, yaitu clustering dan dimensionality reduction. Clustering adalah proses membagi data menjadi beberapa kelompok yang serupa berdasarkan karakteristik yang saling berdekatan. Sedangkan, dimensionality reduction adalah proses mengurangi dimensi dari data sehingga dapat diwakili dalam ruang yang lebih sederhana dan mudah diinterpretasikan.

Dalam unsupervised learning, model tidak diberi label yang jelas, sehingga interpretasi hasilnya bisa sulit. Namun, unsupervised learning sering digunakan untuk eksplorasi data dan identifikasi pola yang menarik, yang dapat membantu peneliti dan praktisi memahami data dengan lebih baik dan menghasilkan wawasan yang berharga. Contoh penggunaan unsupervised learning adalah dalam pengelompokan produk untuk rekomendasi, identifikasi anomali dalam data, atau analisis teks untuk klasifikasi topik.

Penggunaan iris dataset sebagai contoh untuk unsupervised learning memiliki beberapa alasan. Pertama, iris dataset memiliki sedikit fitur, sehingga mudah diinterpretasikan dan divisualisasikan. Kedua, iris dataset memiliki kelompok yang jelas berdasarkan spesies bunga, sehingga hasil dari clustering atau reduksi dimensi dapat langsung dievaluasi dengan menggunakan label asli. Ketiga, iris dataset memiliki distribusi data yang cukup baik dan tidak terlalu rumit, sehingga dapat membantu dalam memahami konsep unsupervised learning tanpa terlalu membingungkan.

Beberapa contoh algoritma unsupervised learning yang sering diterapkan pada iris dataset adalah K-means clustering, hierarchical clustering, PCA (Principal Component Analysis), dan t-SNE (t-Distributed Stochastic Neighbor Embedding). Melalui pengaplikasian algoritma-algoritma tersebut, pengguna dapat memahami bagaimana unsupervised learning dapat digunakan untuk menemukan pola atau struktur yang tersembunyi dalam data tanpa adanya label atau jawaban yang benar sebelumnya.

Score atau skor pada umumnya merujuk pada metrik evaluasi yang digunakan untuk mengukur kinerja model, baik itu dalam supervised learning maupun unsupervised learning. Metrik evaluasi ini biasanya digunakan untuk membandingkan performa model pada dataset yang berbeda, memilih model terbaik, dan melakukan tuning pada hyperparameter.

accuracy_score umumnya digunakan untuk mengevaluasi kinerja model pada tugas klasifikasi supervised learning. Metrik ini mengukur seberapa akurat model dalam memprediksi label atau kelas target pada data uji. accuracy_score akan memberikan nilai antara 0 hingga 1, dimana semakin tinggi skor yang dihasilkan, semakin baik kinerja model tersebut.

Sementara itu, silhouette_score merupakan salah satu metode evaluasi yang digunakan khusus untuk mengevaluasi kualitas clustering pada data unsupervised learning. Silhouette score mengukur seberapa baik setiap data point cocok dengan cluster yang ditempatinya berdasarkan jarak dengan data point yang lain dalam cluster yang sama dibandingkan dengan jarak dengan data point pada cluster lainnya. Score ini memiliki rentang antara -1 hingga 1, dimana semakin dekat dengan 1 (semakin tinggi) artinya clustering semakin baik.

Dalam hal ini, perbedaan utama antara accuracy_score dan silhouette_score adalah pada konteks penggunaannya. accuracy_score digunakan untuk mengevaluasi kinerja model pada tugas klasifikasi supervised learning, sementara silhouette_score digunakan untuk mengevaluasi kualitas clustering pada data unsupervised learning. Oleh karena itu, penggunaan masing-masing metrik ini harus disesuaikan dengan tujuan analisis data yang dilakukan.


Calinski-Harabasz Index atau Variance Ratio Criterion adalah salah satu metrik evaluasi clustering yang digunakan untuk mengukur kualitas cluster yang dihasilkan oleh algoritma clustering. Metrik ini mengukur rasio antara varians antar-cluster dengan varians dalam-cluster, dengan semakin tinggi nilai rasio varians antar-cluster dan semakin rendah nilai rasio varians dalam-cluster menunjukkan semakin baiknya kualitas cluster yang dihasilkan.

Rumus dari Calinski-Harabasz Index adalah:

> CH = (B/(k-1))/(W/(n-k))

dimana:
- B : varians antar-cluster (between-cluster variance)
- W : varians dalam-cluster (within-cluster variance)
- k : jumlah cluster yang dihasilkan oleh algoritma clustering
- n : jumlah data dalam dataset

Semakin tinggi nilai CH, maka semakin baik kualitas cluster yang dihasilkan. Nilai CH dapat dihitung dengan menggunakan fungsi calinski_harabasz_score dari library sklearn.metrics pada Python. Metrik evaluasi ini cocok digunakan pada dataset dengan jumlah cluster yang tidak terlalu banyak, karena semakin banyak cluster yang dihasilkan, semakin tinggi nilai CH yang dihasilkan.

```python
# K-means Clustering

from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
from sklearn.metrics import silhouette_score

# Load the Iris dataset
iris = load_iris()

# Extract the features
X = iris.data
y_true = iris.target

# Initialize the KMeans algorithm
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=0)

# Fit the data to the KMeans algorithm
kmeans.fit(X)

# Get the predicted cluster labels
y_pred = kmeans.predict(X)
labels = kmeans.labels_

# Calculate silhouette score
silhouette_avg = silhouette_score(X, labels)
print("Silhouette Score: ", silhouette_avg)

# Calculate Calinski-Harabasz score
calinski_harabasz_avg = calinski_harabasz_score(X, labels)
print("Calinski-Harabasz Score: ", calinski_harabasz_avg)

# Calculate accuracy score
score = accuracy_score(y_true, y_pred)
print("Accuracy score:", score)
```

```python
# Hierarchical Clustering
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.metrics import accuracy_score
from sklearn.metrics import silhouette_score

# Load iris dataset
iris = load_iris()

# Extract the features
X = iris.data
y_true = iris.target

# Perform hierarchical clustering
model = AgglomerativeClustering(n_clusters=3, linkage='ward')

# Fit the data to the model
model.fit(X)

# Get the predicted cluster labels
labels = model.labels_

# Calculate silhouette score
silhouette_avg = silhouette_score(X, labels)
print("Silhouette Score: ", silhouette_avg)

# Calculate Calinski-Harabasz score
calinski_harabasz_avg = calinski_harabasz_score(X, labels)
print("Calinski-Harabasz Score: ", calinski_harabasz_avg)

# Visualize dendrogram
linkage_matrix = linkage(X, 'ward')
dendrogram(linkage_matrix)

plt.title('Dendrogram')
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.show()
```

```python
# Hierarchial Clustering dengan Scaling
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.metrics import accuracy_score
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler

# Load iris dataset
iris = load_iris()

# Extract the features
X = iris.data
y_true = iris.target

# Perform scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform hierarchical clustering
model = AgglomerativeClustering(n_clusters=3, linkage='ward')

# Fit the data to the model
model.fit(X_scaled)

# Get the predicted cluster labels
labels = model.labels_

# Calculate silhouette score
silhouette_avg = silhouette_score(X_scaled, labels)
print("Silhouette Score: ", silhouette_avg)

# Calculate Calinski-Harabasz score
calinski_harabasz_avg = calinski_harabasz_score(X_scaled, labels)
print("Calinski-Harabasz Score: ", calinski_harabasz_avg)

# Visualize dendrogram
linkage_matrix = linkage(X_scaled, 'ward')
dendrogram(linkage_matrix)

plt.title('Dendrogram')
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.show()
```

# Principal Component Analysis (PCA)

PCA adalah sebuah teknik reduksi dimensi pada data. Tujuan utama dari PCA adalah untuk mengidentifikasi pola yang tersembunyi atau fitur-fitur penting dari data yang memiliki dimensi yang lebih tinggi. Dalam banyak kasus, data dengan dimensi yang lebih tinggi akan sulit untuk dianalisis dan divisualisasikan.

PCA bekerja dengan cara memproyeksikan data yang berada di dalam sebuah ruang dengan dimensi yang tinggi ke dalam sebuah ruang dengan dimensi yang lebih rendah, sehingga fitur-fitur penting yang terkandung dalam data dapat terlihat lebih jelas. Proses pengurangan dimensi ini dilakukan dengan cara mencari kombinasi linier dari fitur-fitur asli yang memiliki variansi yang paling besar, sehingga informasi yang hilang akibat pengurangan dimensi tersebut bisa diminimalkan.

Secara matematis, PCA menghitung nilai eigen dari matriks kovarian antara fitur-fitur asli, dan kemudian menentukan beberapa vektor eigen utama yang merupakan representasi dari fitur-fitur yang memiliki variansi yang paling besar. Vektor eigen utama tersebut akan membentuk sebuah basis baru untuk data, yang kemudian digunakan untuk merepresentasikan data dalam sebuah ruang dengan dimensi yang lebih rendah.

PCA banyak digunakan dalam berbagai aplikasi seperti pengenalan pola, analisis citra, analisis teks, dan masih banyak lagi. Salah satu keuntungan dari penggunaan PCA adalah dapat menghilangkan fitur-fitur yang tidak penting dan mempercepat waktu komputasi pada data yang berukuran besar.

Pada data iris, PCA dapat digunakan untuk mengurangi dimensi dari empat fitur (panjang dan lebar kelopak dan mahkota bunga) menjadi dua fitur utama yang paling banyak memberikan informasi tentang variasi dalam data.

```python
# PCA
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score, silhouette_score, calinski_harabasz_score

# Load the Iris dataset
iris = load_iris()

# Extract the features
X = iris.data
y_true = iris.target

# Apply PCA to reduce the dimensionality
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)

# Initialize the KMeans algorithm
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=0)

# Fit the reduced data to the KMeans algorithm
kmeans.fit(X_reduced)

# Get the predicted cluster labels
y_pred = kmeans.predict(X_reduced)
labels = kmeans.labels_

# Calculate silhouette score
silhouette_avg = silhouette_score(X, labels)
print("Silhouette Score: ", silhouette_avg)

# Calculate Calinski-Harabasz score
calinski_harabasz_avg = calinski_harabasz_score(X, labels)
print("Calinski-Harabasz Score: ", calinski_harabasz_avg)

# Calculate accuracy score
score = accuracy_score(y_true, y_pred)
print("Accuracy score:", score)
```

```python
# t-SNE
import numpy as np
from sklearn.datasets import load_iris
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score, silhouette_score, calinski_harabasz_score

# Load iris dataset
iris = load_iris()

# Extract the features
X = iris.data
y_true = iris.target

# Perform t-SNE
tsne = TSNE(n_components=2, perplexity=30, learning_rate=200)
X_tsne = tsne.fit_transform(X)

# Perform k-means clustering on t-SNE results
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=0)
kmeans.fit(X_tsne)

y_pred = kmeans.predict(X_tsne)
labels = kmeans.labels_

# Calculate silhouette score
silhouette_avg = silhouette_score(X_tsne, labels)
print("Silhouette Score: ", silhouette_avg)

# Calculate Calinski-Harabasz score
calinski_harabasz_avg = calinski_harabasz_score(X_tsne, labels)
print("Calinski-Harabasz Score: ", calinski_harabasz_avg)

# Calculate accuracy score
score = accuracy_score(y_true, y_pred)
print("Accuracy score:", score)
```

# Scree Plot pada PCA



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

# Load Iris dataset
iris = load_iris()
X = iris.data

# Fit PCA to the data
pca = PCA().fit(X)

# Plot scree plot
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Scree Plot')
plt.show()
```

# Scaling
Scaling sebelum melakukan PCA sangat penting untuk memastikan bahwa semua fitur (variabel) memiliki skala yang sama, sehingga tidak ada variabel yang mendominasi atau memiliki bobot yang lebih besar daripada yang lain. Hal ini sangat penting karena PCA didasarkan pada kovarian antar fitur, sehingga fitur dengan skala yang besar dapat memiliki pengaruh yang lebih besar dalam membangun principal component.

Dengan melakukan scaling pada data sebelum PCA, kita dapat menormalkan data dan membuat variabel-variabel memiliki varian yang serupa, sehingga mereka akan memiliki kontribusi yang seimbang dalam membangun principal component. Tanpa scaling, hasil dari PCA mungkin akan terpengaruh oleh skala data, sehingga menghasilkan principal component yang salah dan kurang akurat.

Selain itu, scaling juga dapat membantu mengurangi noise dan variabilitas data, sehingga PCA dapat bekerja lebih efektif. Oleh karena itu, sangat disarankan untuk melakukan scaling pada data sebelum melakukan PCA. Beberapa teknik scaling yang umum digunakan sebelum melakukan PCA antara lain Standard Scaler, Min-Max Scaler, dan Robust Scaler.

```python
# Tanpa Scaling
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

# Load iris dataset
iris = load_iris()

# Extract the features
X = iris.data

# Perform PCA without scaling
pca = PCA()
pca.fit(X)

# Plot scree plot
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance')
plt.title('Scree plot without scaling')
plt.show()
```

```python
# Dengan Scaling
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Load iris dataset
iris = load_iris()

# Extract the features
X = iris.data

# Perform scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform PCA with scaling
pca = PCA()
pca.fit(X_scaled)

# Plot scree plot
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance')
plt.title('Scree plot with scaling')
plt.show()
```

```python
# Tanpa Scaling

from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

iris = load_iris()
X = iris.data

pca = PCA()
pca.fit(X)
eigenvalues = pca.explained_variance_

plt.plot(range(1, len(eigenvalues) + 1), eigenvalues, 'bo-', linewidth=2)
plt.title('Scree Plot Tanpa Scaling')
plt.xlabel('Principal Component')
plt.ylabel('Eigenvalue')
plt.show()
```
